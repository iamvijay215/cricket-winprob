## Data Pre-Processing

Goal: Make raw data usable for analysis/modeling.

Happens before modeling and often guided by insights from EDA.

Itâ€™s about cleaning and standardizing.

Typical steps:

    - Handling missing values (NaN, blanks).

    - Removing duplicates.

    - Converting data types (string â†’ datetime, object â†’ numeric).

    - Encoding categorical variables (one-hot, label encoding).

    - Normalization/standardization of numerical features.

    - Splitting data into train/test sets.

âš™ï¸ Think of preprocessing as housekeeping: youâ€™re tidying the data so algorithms wonâ€™t choke.

### Data Cleaning for Matches.csv

```python
# 1) Basic overview
matches.head()
matches.info()
matches.isna().sum()

# 2) Type fixes
matches["date"] = pd.to_datetime(matches["date"], errors="coerce")
matches["dl_applied"] = matches["dl_applied"].astype("int64")  # keep as 0/1
```

```python
matches.info()                                    # dtypes + Non-Null counts
matches.isna().sum().sort_values(ascending=False) # how many NaNs per column
```

```python
cols_to_drop = [c for c in ["umpire3"] if c in matches.columns]  # only drop if present
matches = matches.drop(columns=cols_to_drop)                      # remove useless empty cols
```

```python
matches.head(5)
```

```python
# Convert 'date' text â†’ datetime; invalid values become NaT (safe)
matches["date"] = pd.to_datetime(matches["date"], errors="coerce")

# Ensure DL flag is integer 0/1 (sometimes comes as object)
matches["dl_applied"] = matches["dl_applied"].astype("int64")
```

```python
matches.head(3)
```

```python
# Strip leading/trailing spaces and normalize multiple spaces â†’ single space
for col in ["team1", "team2", "toss_winner", "winner", "venue", "city", "toss_decision", "result"]:
    matches[col] = matches[col].astype(str).str.strip().str.replace(r"\s+", " ", regex=True)
```

```python
matches.head(3)
```

```python
# Keep historical names as-is (no renaming to today's franchises), just fix known typos
team_fix = {
    "Rising Pune Supergiants": "Rising Pune Supergiant",  # Kaggle inconsistency
    # add more if you see them during EDA
}
for col in ["team1", "team2", "toss_winner", "winner"]:
    matches[col] = matches[col].replace(team_fix)
```

```python
matches.head(3)
```

```python
# Fill missing city intelligently from venue (safe mapping)

# Build a venueâ†’city map from rows that already have both values
v2c = (matches[~matches["city"].isna()][["venue", "city"]]
       .drop_duplicates()
       .set_index("venue")["city"].to_dict())
```

```python
matches.head(3)
```

```python
# For rows with city missing, try to fill from the venueâ†’city map
mask_city_na = matches["city"].isna()
matches.loc[mask_city_na, "city"] = matches.loc[mask_city_na, "venue"].map(v2c)

# If any city still missing (rare), fill with "Unknown" (or keep NaN if you prefer)
matches["city"] = matches["city"].fillna("Unknown")
```

```python
matches.head(3)
```

```python
# Winner / player_of_match consistency for â€œno resultâ€

# If result is 'no result', winner and player_of_match are expected to be NaN â†’ keep as-is.
# If you prefer explicit labels (optional), uncomment:
# nores = matches["result"].str.lower().eq("no result")
# matches.loc[nores & matches["winner"].isna(), "winner"] = "No Result"
# matches.loc[nores & matches["player_of_match"].isna(), "player_of_match"] = "None"
```

```python
 # Sanity checks (duplicates, remaining NaNs)
    
# Match ID should be unique
dup_ids = matches.duplicated(subset=["id"]).sum()
print("Duplicate match IDs:", dup_ids)            # expect 0

# Remaining missing values by column
matches.isna().sum().sort_values(ascending=False)
```

```python
from pathlib import Path

# Define output path for cleaned matches file
OUT = Path("../data/processed/matches_clean.csv")

# Ensure directory exists
OUT.parent.mkdir(parents=True, exist_ok=True)

# Save cleaned dataset
matches.to_csv(OUT, index=False)
print(f"Saved cleaned matches â†’ {OUT}")
```

### CLeaning the Deliveries.csv

```python
from pathlib import Path
import pandas as pd
import numpy as np

RAW_DEL = Path("../data/raw/ipl/deliveries.csv")          # raw deliveries
OUT_DEL = Path("../data/processed/deliveries_clean.csv")  # cleaned output

deliv = pd.read_csv(RAW_DEL)                               # load
print(deliv.shape)                                         # (rows, cols)
deliv.head(3)                                              # quick peek
```

```python
deliv.info()                                               # dtypes + non-nulls
deliv.isna().sum().sort_values(ascending=False)            # missing values per column
```

```python
# Strip whitespace in text columns (safe normalization)

text_cols = ["batting_team","bowling_team","batsman","non_striker","bowler",
             "player_dismissed","dismissal_kind","fielder"]
for c in text_cols:
    deliv[c] = deliv[c].astype(str).str.strip().replace({"nan": np.nan})
```

```python
# Team name standardization (same as matches)

team_fix = {
    "Rising Pune Supergiants": "Rising Pune Supergiant",
    # add more if you spot them during EDA
}
for c in ["batting_team","bowling_team"]:
    deliv[c] = deliv[c].replace(team_fix)
```

```python
# Type fixes for numeric columns

int_like = ["match_id","inning","over","ball","is_super_over",
            "wide_runs","bye_runs","legbye_runs","noball_runs","penalty_runs",
            "batsman_runs","extra_runs","total_runs"]
for c in int_like:
    deliv[c] = pd.to_numeric(deliv[c], errors="coerce").astype("Int64")  # nullable ints

# quick sanity after coercion
deliv[int_like].isna().sum().sort_values(ascending=False)
```

```python
deliv.head(3)
```

```python
deliv.isna().sum().sort_values(ascending=False)
```

```python
# If there are any NaNs after coercion, that means the raw file had bad valuesâ€”usually rare. so, dropping such corrupted rows carefully:

bad_numeric = deliv[int_like].isna().any(axis=1)
print("Rows with invalid numeric values:", bad_numeric.sum())
deliv = deliv.loc[~bad_numeric].copy()
for c in int_like:
    deliv[c] = deliv[c].astype("int64")  # back to regular int after drop
```

```python
# Logical consistency checks and fixes

# A) Extras + batsman_runs should equal total_runs

bad_totals = deliv[(deliv["batsman_runs"] + deliv["extra_runs"]) != deliv["total_runs"]]
print("Rows with inconsistent totals:", len(bad_totals))
# If any appear (rare), recompute total_runs from components:
deliv["total_runs"] = deliv["batsman_runs"] + deliv["extra_runs"]

# B) No negative runs anywhere

num_cols = ["wide_runs","bye_runs","legbye_runs","noball_runs",
            "penalty_runs","batsman_runs","extra_runs","total_runs"]
neg_rows = (deliv[num_cols] < 0).any(axis=1)
print("Rows with negative values:", neg_rows.sum())
deliv = deliv.loc[~neg_rows].copy()


# C) Over/ball range sanity

# over should be 1..20 (super overs are flagged separately)
bad_over = ~deliv["over"].between(1, 20)
print("Rows with invalid over number:", bad_over.sum())
deliv = deliv.loc[~bad_over].copy()

# ball should be >=1 (can exceed 6 due to wides/no-balls that don't count)
bad_ball = deliv["ball"] < 1
print("Rows with invalid ball number:", bad_ball.sum())
deliv = deliv.loc[~bad_ball].copy()
```

```python
# Wicket flag for easier modeling (helper feature; optional to keep)

# If player_dismissed has a name, that ball had a wicket
deliv["is_wicket"] = deliv["player_dismissed"].notna().astype("int8")
deliv["dismissal_kind"] = deliv["dismissal_kind"].fillna("not_out")  # clearer label for later grouping
```

```python
# Integrity with matches (foreign-key check)

# Load cleaned matches ids (ensure you saved matches_clean.csv first)
matches_clean = pd.read_csv("../data/processed/matches_clean.csv", usecols=["id"])
valid_ids = set(matches_clean["id"].unique())

missing_fk = ~deliv["match_id"].isin(valid_ids)
print("Deliveries with unknown match_id:", missing_fk.sum())
deliv = deliv.loc[~missing_fk].copy()
```

```python
# Final null review (should be only in optional text fields)

deliv.isna().sum().sort_values(ascending=False).head(10)
```

```python
# Save cleaned deliveries

OUT_DEL.parent.mkdir(parents=True, exist_ok=True)
deliv.to_csv(OUT_DEL, index=False)
print(f"Saved cleaned deliveries â†’ {OUT_DEL}")
```
